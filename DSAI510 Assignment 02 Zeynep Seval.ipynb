## Problem 3 (10 pts)
- The file **EV_prices.csv** contains prices for various makes, models, and years of cars. Load this file into a dataframe. Rename the column **Model Year** to **Year**.
- We want to add a new column **Price** to our dataframe from the previous problem. This column will include the price of the car for the corresponding make, model and year if this information is available in the file **EV_prices.csv**. If not available, we'll still keep the record but the entry for price will be empty, NA, None or NaN. To achieve this, merge the dataframe from the previous problem with the dataframe containing the data from  **EV_prices.csv**. Think carefully and decide if you need to merge with 'inner' or 'outer' method. At the end, we should have these columns in the merged dataframe: **Year**, **Make**, **Model**, **Electric Range**, **Latitude**, **Longitude** and **Price**. Again, the **Price** column will have numbers only for some records, but it will be empty or NaN for the rest.
- Next, show the number of records which has price information in the **Price** column. Hint: You can use a one-liner containing `len()` and `dropna()` together.

```python
import pandas as pd
import numpy as np

# Load the EV_prices.csv file
prices_df = pd.read_csv(r"C:\Users\seval\OneDrive\Desktop\DSAI 510\HW2\EV_prices.csv")

# Rename the 'Model Year' column to 'Year' in prices_df
prices_df = prices_df.rename(columns={'Model Year': 'Year'})

# Assuming df_new is our dataframe from the previous problem
# Perform a left merge of df_new and prices_df
# We use a left merge because:
# 1. We want to keep all vehicle records from our original dataset (df_new)
# 2. We want to add price information only where available
# 3. We don't want to introduce new vehicles that might be in prices_df but not in df_new
# This ensures we maintain all our original data while enriching it with prices where possible
merged_df = pd.merge(df_new, prices_df, on=['Year', 'Make', 'Model'], how='left')

# Select only the required columns
final_df = merged_df[['Year', 'Make', 'Model', 'Electric Range', 'Latitude', 'Longitude', 'Price']]

# Count the number of records with price information
# This one-liner drops all rows where 'Price' is NaN and then counts the remaining rows
records_with_price = len(final_df.dropna(subset=['Price']))

print(f"Number of records with price information: {records_with_price}")

# Display the first few rows of the final dataframe
print("\nFirst few rows of the final dataframe:")
print(final_df.head())

# Display dataframe info
print("\nDataframe info:")
print(final_df.info())
```

```python
import pandas as pd
import numpy as np
import warnings

# Suppress warnings about performance with large datasets
warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)

# Assuming final_df is our DataFrame from the previous problem

# Step 1: Remove records where Year is 2015 or earlier
final_df = final_df[final_df['Year'] > 2015]

# Step 2: Create the pivot table using pivot_table() method and size aggregation
pivot_table = pd.pivot_table(
    final_df,  # The DataFrame we're working with
    values='Electric Range',  # The column used for aggregation. With 'size', the choice doesn't matter
    index=['Make', 'Model'],  # Creates a multi-index with Make and Model
    columns='Year',  # Spreads out the Years as columns
    aggfunc='size',  # 'size' counts the number of entries for each Make/Model/Year combination
    fill_value=0  # Replaces NaN with 0 for combinations that don't exist in the data
)
# The resulting pivot_table will have:
# - Rows indexed by (Make, Model)
# - Columns representing each Year
# - Cell values showing the count of cars for each Make/Model/Year combination

# Ensure all years from 2016 to 2024 are present
all_years = range(2016, 2025)
for year in all_years:
    if year not in pivot_table.columns:
        pivot_table[year] = 0  # Add missing years with count 0

# Sort columns and convert to float
pivot_table = pivot_table.reindex(columns=sorted(pivot_table.columns)).astype(float)

# Display the pivot table
print("Pivot Table (first 20 rows):")
display(pivot_table.head(21))
```

```python
# Assuming final_df is our DataFrame from the previous problem

# Step 1: Remove records where Year is 2015 or earlier
final_df = final_df[final_df['Year'] > 2015]

# Step 2: Use groupby to calculate average latitude and round to 2 decimal places
grouped_df = final_df.groupby(['Make', 'Model', 'Year'])['Latitude'].mean().round(2).reset_index()

# Step 3: Pivot the result to create a table similar to the previous one
avg_latitude_table = grouped_df.pivot(index=['Make', 'Model'], columns='Year', values='Latitude')

# Step 4: Ensure all years from 2016 to 2024 are present
all_years = range(2016, 2025)
for year in all_years:
    if year not in avg_latitude_table.columns:
        avg_latitude_table[year] = np.nan

# Step 5: Sort columns and fill NaN with a placeholder (e.g., 0.00)
avg_latitude_table = avg_latitude_table.reindex(columns=sorted(avg_latitude_table.columns)).fillna(0.00)

# Display the average latitude table
print("Average Latitude Table (first 20 rows):")
display(avg_latitude_table.head(21))
```

```python
# Import the required library
import pandas as pd

# Create the initial dataset
data = {
    'student_id': [1, 2, 3],
    'Math_Q1': [85, 90, 82],
    'Math_Q2': [88, 85, 80],
    'Math_Q3': [87, 83, 84],
    'Science_Q1': [78, 88, 80],
    'Science_Q2': [82, 85, 78],
    'Science_Q3': [84, 87, 83]
}

# Create a DataFrame from the data
scores_df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(scores_df)
print("\n")  # Add a blank line for better readability

# Use melt() to reshape the DataFrame
melted_df = pd.melt(
    scores_df,
    # Specify the column to keep as is
    id_vars=['student_id'],
    # Specify the columns to "melt" (all except 'student_id')
    value_vars=['Math_Q1', 'Math_Q2', 'Math_Q3', 'Science_Q1', 'Science_Q2', 'Science_Q3'],
    # Name the new column for the melted column names
    var_name='Subject_Quarter',
    # Name the new column for the values
    value_name='Score'
)

# Sort the melted DataFrame by student_id and Subject_Quarter
melted_df = melted_df.sort_values(['student_id', 'Subject_Quarter'])

# Reset the index to start from 0
melted_df = melted_df.reset_index(drop=True)

# Display the melted DataFrame
print("Melted DataFrame:")
print(melted_df)
```

## Problem 5 - Quality Control in a Manufacturing Plant (10 pts)

Imagine you work as a quality control analyst in a manufacturing plant that produces ball bearings. Each day, multiple batches of ball bearings are produced. To ensure the consistency and quality of the ball bearings, samples from each batch are measured to determine their diameters.

Over the course of a month, you've collected diameter data for these samples from various batches. The objective is to determine the batch consistency by measuring the standard deviation of the diameters. A lower standard deviation would indicate that the ball bearings in a batch are more consistent in size.

- Load the **ball_bearings.csv** file into a dataframe.
- Use `groupby()` to calculate the standard deviation for each batch.
- Sort the results in descending order wrt standard deviation, showing the batch with highest standard deviation at the top.



```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv(r"C:\Users\seval\OneDrive\Desktop\DSAI 510\HW2\ball_bearings.csv")

# Display the first few rows of the loaded data to verify
print("First few rows of the loaded data:")
print(df.head())
print("\n")  # Add a blank line for better readability
```

```python
# Group the data by batch_id and calculate the standard deviation of diameters
batch_std = df.groupby('batch_id')['diameter'].agg(['std'])

# Sort the results in descending order of standard deviation
batch_std_sorted = batch_std.sort_values('std', ascending=False)

# Display the results
print("Batch Standard Deviations (Sorted in Descending Order):")
print(batch_std_sorted)
```
